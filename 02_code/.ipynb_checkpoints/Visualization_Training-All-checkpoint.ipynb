# Preprocessing

```python
df_predictor = df[['hospital_death']]
df_predictor.head()
```

```python
## return only numeric columns from dataframe
df_numeric = df._get_numeric_data().drop(['encounter_id'], axis = 1)
df_numeric.head(200)
```

```python
print(df_numeric.columns.values)
```

```python
df_categorical =df.drop(df_numeric, axis = 1).drop(['encounter_id'], axis = 1)
df_categorical['hospital_death'] = df_predictor['hospital_death']
df_categorical.head()
```

```python
sns.set(style="ticks")
fig, ax = plt.subplots(3, 3, figsize = (20, 20))
for cat, subplot in zip(df_categorical.columns.values, ax.flatten()):
    sns.countplot(df_categorical[cat],hue = df_categorical['hospital_death'],ax = subplot, data = df_categorical)
    plt.subplots_adjust(hspace = 0.75)
    for label in subplot.get_xticklabels():
        label.set_rotation(90)
        
fig.savefig('../03_figures/Categorical_Countplot'  + (str(date.today())))

    
```

```python
#How many patients in this dataset live and died?
fig, axs = plt.subplots(1,2,figsize=(14,7))
sns.countplot(x='hospital_death', data = df, ax = axs[0])
axs[0].set_title("Frequency of Survival Status")
df['hospital_death'].value_counts().plot(x=None, y = None, kind = 'pie', ax = axs[1],autopct='%1.2f%%')
axs[1].set_title("Percentage of Survival Status")
plt.show()
```

```python
# Handle Missing Values
df_categorical.shape
```

```python
#Nulls
null_counts = df_categorical.isnull().sum()
print("Number of null values in each column:\n{}".format(null_counts))
```

```python
new_df = df[['encounter_id','hospital_death','apache_3j_bodysystem', 'apache_2_bodysystem']]
```

```python
new_df.shape
```

```python
new_df.head()
```

```python
#Are we working with all numeric data?
print("Data types and their frequency\n{}".format(new_df.dtypes.value_counts()))
```

```python
#drop rows with Null values
new_df = new_df.dropna(axis = 0)
null_counts = new_df.isnull().sum()
print("Number of null values in each column:\n{}".format(null_counts))



new_df.shape
```

```python
# #label Encoding
# le = LabelEncoder()
# #Apply encoder
# le_df = new_df.apply(le.fit_transform)
# le_df.head()
```

```python
#Get Dummies
enc_df = pd.get_dummies(new_df[2:], drop_first = True)
enc_df.head()
```

```python
#TRAINING
```

```python
# #Export Numeric Train Data
enc_df.to_csv("../01_data/training_CategoricalOnly.csv", index = False)

#combine categorical to the numeric 
#Import Data
dfpath = "../01_data/training_NumericOnly.csv"
df_num = pd.read_csv(dfpath)
df_num.head()


final_df = pd.merge(enc_df, df_num, on = ['encounter_id','hospital_death'])
final_df.head()
final_df.shape

#Export Numeric Train Data
final_df.to_csv("../01_data/training_catANDnum.csv", index = False)
```
