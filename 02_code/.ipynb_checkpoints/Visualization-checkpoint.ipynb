```python
# Handle Missing Values
df.shape
```

```python
#Nulls
null_counts = df.isnull().sum()
print("Number of null values in each column:\n{}".format(null_counts))
```

```python
#drop columns with > 25% of rows with missing values  - How does this impact our attributes?
threshold = int(len(df) * .25)
print(threshold)
df = df.dropna(thresh = threshold, axis = 1)
df.head()
    
```

```python
df.shape
```

```python
#drop all rows with missin values?  How small will our dataset be?
dfCleaned = df.dropna(axis = 0)
```

```python
dfCleaned.shape
```

```python
#How many patients in this dataset live and died?
fig, axs = plt.subplots(1,2,figsize=(14,7))
sns.countplot(x='hospital_death', data = dfCleaned, ax = axs[0])
axs[0].set_title("Frequency of Survival Status")
dfCleaned['hospital_death'].value_counts().plot(x=None, y = None, kind = 'pie', ax = axs[1],autopct='%1.2f%%')
axs[1].set_title("Percentage of Survival Status")
plt.show()
```
